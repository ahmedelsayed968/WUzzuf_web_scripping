```python

# Press Shift+F10 to execute it or replace it with your code.
# Press Double Shift to search everywhere for classes, files, tool windows, actions, and settings.

# from selenium import webdriver
# from selenium.webdriver.common.keys import Keys
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

class Wuzzuf:
    def __init__(self):
        self.job_titles = []
        self.company_names = []
        self.locations = []
        self.durations = []
        self.Experiences = []
        self.soup = None
    def get_jobs_from_wuzzuf(self, job_name: str, page_num: int):
        
        if page_num >= 1:
            url_base = 'https://wuzzuf.net/search/jobs/?a=hpb&'
            job_search_url = url_base + 'q={}'.format(job_name.replace(' ', '%20'))
            respones = requests.get(job_search_url)
            if respones.ok:
                for num in range(page_num):
                    url_page = job_search_url + '&start={}'.format(num)
                    respone_page = requests.get(url_page)
                    print(url_page)
                    if respone_page.ok:
                        self.soup = BeautifulSoup(respone_page.text)
                        self.__get_job_titles()
                        time.sleep(2)
                        self.__get_company_names()
                        time.sleep(2)
                        self.__get_locations()
                        time.sleep(2)
                        self.__get_durations()
                        time.sleep(2)
                        self.__get_experiences()
                        time.sleep(4)
                    else:
                        print('Error ' + respone_page.status_code)
                        return
                self.__filter_duration()    
                map_return = {'job_titles': self.job_titles,
                              'company_name': self.company_names,
                              'location': self.locations,
                              'duration': self.durations,
                              'Experience': self.Experiences}
                return map_return
            else:
                print('Erorr ' + respones.status_code)
                return None
        else:
            return None

    def __get_job_titles(self) -> None:
        for item in self.soup.findAll('a', attrs={'target': '_blank', 'rel': 'noreferrer', 'class': 'css-o171kl'}):
            self.job_titles.append(item.text)

    def __get_company_names(self) -> None:
        search = self.soup.find_all('a', {'class': 'css-17s97q8'})
        for company in search:
            self.company_names.append(company.text)

    def __get_locations(self) -> None:
        location_search = self.soup.find_all('span', attrs={'class': 'css-5wys0k'})
        for location in location_search:
            self.locations.append(location.text)

    def __get_durations(self) -> None:
        div_dur = self.soup.find_all('div', attrs={'class': 'css-1lh32fc'})
        for div in div_dur:
            dur = div.find_all('span')
            temp = []
            for item in dur:
                temp.append(item.text)
            self.durations.append(temp)

    def __filter_duration(self) -> None:
        for idx in range(len(self.durations)):
            self.durations[idx] = str(','.join(self.durations[idx]))

    def __get_experiences(self) -> None:
        ll = self.soup.find_all('div', attrs={'class': None})

        for idex in range(6, len(ll)):
            if ll[idex].span is not None:
                text = ll[idex].span.text
                if 'exp' in text.lower():
                    self.Experiences.append(text)
                else:
                    self.Experiences.append('Internship')
            else:
                self.Experiences.append(np.nan)

    def __del__(self):
        print('deleted')


def main() -> None:
       pass
if __name__ == '__main__':
    main()

# See PyCharm help at https://www.jetbrains.com/help/pycharm/
```
